```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib ipympl
```

```python
day_ahead_Jan = pd.read_csv('day_ahead_data_Jan.csv')
day_ahead_Aug = pd.read_csv('day_ahead_data_Aug.csv')
day_ahead_Oct = pd.read_csv('day_ahead_data_Oct.csv')
real_time_Jan = pd.read_csv('real_time_data_Jan.csv')
real_time_Aug = pd.read_csv('real_time_data_Aug.csv')
real_time_Oct = pd.read_csv('real_time_data_Oct.csv')
```

```python
node_list = day_ahead_Jan['Location'].unique()
list(node_list)

```

```python
"""set the date column as date time format"""
real_time_Jan['Time'] = pd.to_datetime(real_time_Jan['Time'])
"""get the the rows where date is Jan 5"""
real_time_Jan_5 = real_time_Jan[real_time_Jan['Time'].dt.day == 5]
print(real_time_Jan_5[real_time_Jan_5['Location'] == 'VACA-DIX_1_B1' ]['Time'])
```

```python
real_time_Jan = real_time_Jan[real_time_Jan['Time'].dt.day != 5]
```

```python
day_ahead_Jan['Time'] = pd.to_datetime(day_ahead_Jan['Time'])
"""remove the rows where date is Jan 5"""
day_ahead_Jan = day_ahead_Jan[day_ahead_Jan['Time'].dt.day != 5]    
day_ahead_Jan
```

```python
real_time_Aug['Time'] = pd.to_datetime(real_time_Aug['Time'])
sucker = real_time_Aug[real_time_Aug['Location'] == 'WISE1_7_B1']
"""groupby date and count the number of rows for each date"""
sucker.groupby(sucker['Time'].dt.date).count()

```

```python
"""renmove the rows where date is Aug 1, Aug 24, and Aug 25"""
real_time_Aug = real_time_Aug[real_time_Aug['Time'].dt.day != 1]
real_time_Aug = real_time_Aug[real_time_Aug['Time'].dt.day != 24]
real_time_Aug = real_time_Aug[real_time_Aug['Time'].dt.day != 25]
day_ahead_Aug['Time'] = pd.to_datetime(day_ahead_Aug['Time'])
"""remove the rows where date is Aug 1, Aug 24, and Aug 25"""
day_ahead_Aug = day_ahead_Aug[day_ahead_Aug['Time'].dt.day != 1]
day_ahead_Aug = day_ahead_Aug[day_ahead_Aug['Time'].dt.day != 24]
day_ahead_Aug = day_ahead_Aug[day_ahead_Aug['Time'].dt.day != 25]
```

```python
print(len(day_ahead_Aug[day_ahead_Aug['Location'] == 'WISE1_7_B1']))
print(len(real_time_Aug[real_time_Aug['Location'] == 'WISE1_7_B1']))
```

```python
real_time_Oct['Time'] = pd.to_datetime(real_time_Oct['Time'])
doomer = real_time_Oct[real_time_Oct['Location'] == 'WISE1_7_B1']
"""groupby date and count the number of rows for each date"""
doomer.groupby(doomer['Time'].dt.date).count()
```

```python
"""remove the Oct 17 data"""
real_time_Oct = real_time_Oct[real_time_Oct['Time'].dt.day != 17]
day_ahead_Oct['Time'] = pd.to_datetime(day_ahead_Oct['Time'])
"""remove the Oct 17 data"""
day_ahead_Oct = day_ahead_Oct[day_ahead_Oct['Time'].dt.day != 17]
```

```python
print(len(day_ahead_Oct[day_ahead_Oct['Location'] == 'WISE1_7_B1']))
print(len(real_time_Oct[real_time_Oct['Location'] == 'WISE1_7_B1']))
```

```python
table_real_time_Jan = dict()
for node in node_list:
    a = real_time_Jan[real_time_Jan['Location'] == node]
    new_column = np.array([])
    for i in range(int(len(a) / 4)):
        new_column = np.append(new_column, np.mean(a['LMP'][i*4:(i+1)*4]))

    table_real_time_Jan[node] = new_column

    
    
```

```python
table_real_time_Aug = dict()
for node in node_list:
    a = real_time_Aug[real_time_Aug['Location'] == node]
    new_column = np.array([])
    for i in range(int(len(a) / 4)):
        new_column = np.append(new_column, np.mean(a['LMP'][i*4:(i+1)*4]))

    table_real_time_Aug[node] = new_column
```

```python
table_real_time_Oct = dict()
for node in node_list:
    a = real_time_Oct[real_time_Oct['Location'] == node]
    new_column = np.array([])
    for i in range(int(len(a) / 4)):
        new_column = np.append(new_column, np.mean(a['LMP'][i*4:(i+1)*4]))

    table_real_time_Oct[node] = new_column
```

```python
table_day_ahead_Jan = dict()
for node in node_list:
    a = day_ahead_Jan[day_ahead_Jan['Location'] == node]
    new_column = a['LMP'].to_numpy()
    table_day_ahead_Jan[node] = new_column

table_day_ahead_Aug = dict()
for node in node_list:
    a = day_ahead_Aug[day_ahead_Aug['Location'] == node]
    new_column = a['LMP'].to_numpy()

    table_day_ahead_Aug[node] = new_column

table_day_ahead_Oct = dict()
for node in node_list:
    a = day_ahead_Oct[day_ahead_Oct['Location'] == node]
    new_column = a['LMP'].to_numpy()

    table_day_ahead_Oct[node] = new_column
```

```python
print(len(table_real_time_Jan['VACA-DIX_1_B1']))
print(len(table_day_ahead_Jan['VACA-DIX_1_B1']))
print(len(table_real_time_Aug['WISE1_7_B1']))
print(len(table_day_ahead_Aug['WISE1_7_B1']))
print(len(table_real_time_Oct['WISE1_7_B1']))
print(len(table_day_ahead_Oct['WISE1_7_B1']))

```

```python
table_Jan_diff = dict()
table_Aug_diff = dict()
table_Oct_diff = dict()
for node in node_list:
    table_Jan_diff[node] = table_real_time_Jan[node][24:] - table_day_ahead_Jan[node][:-24]
    table_Aug_diff[node] = table_real_time_Aug[node][24:] - table_day_ahead_Aug[node][:-24]
    table_Oct_diff[node] = table_real_time_Oct[node][24:] - table_day_ahead_Oct[node][:-24]
    
```

```python
print(len(table_Aug_diff['WISE1_7_B1']))
print(len(table_Oct_diff['WISE1_7_B1']))
print(len(table_Jan_diff['VACA-DIX_1_B1']))
```

```python
df_Jan_diff = pd.DataFrame(table_Jan_diff)
df_Aug_diff = pd.DataFrame(table_Aug_diff)
df_Oct_diff = pd.DataFrame(table_Oct_diff)
df_Jan_diff.to_csv('Jan_diff.csv')
df_Aug_diff.to_csv('Aug_diff.csv')
df_Oct_diff.to_csv('Oct_diff.csv')

```
