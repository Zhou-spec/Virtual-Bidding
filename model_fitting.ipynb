```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib ipympl
```

```python
node_list = ['LBEACH1G_7_N001','BIGCRK3_2_B1','MILPITAS_1_N001','IBM-CTLE_1_N002','MECCTG2_7_B2','VACA-DIX_1_B1','KANAKAJT_1_B1','ALTA1G_7_B1','RENFRO_1_N001']
```

```python
df1 = pd.read_csv(node_list[0] + 'Jan.csv')
df1['Node_Name'] = node_list[0]
for node in node_list[1:]:
    df2 = pd.read_csv(node + 'Jan.csv')
    df2['Node_Name'] = node
    df1 = pd.concat([df1,df2], axis = 0)

df_Jan = df1.copy()
```

```python
df1 = pd.read_csv(node_list[0] + 'Aug.csv')
df1['Node_Name'] = node_list[0]
df1
for node in node_list[1:]:
    df2 = pd.read_csv(node + 'Aug.csv')
    df2['Node_Name'] = node
    df1 = pd.concat([df1,df2], axis = 0)

df_Aug = df1.copy()
```

```python
df1 = pd.read_csv(node_list[0] + 'Oct.csv')
df1['Node_Name'] = node_list[0]
for node in node_list[1:]:
    df2 = pd.read_csv(node + 'Oct.csv')
    df2['Node_Name'] = node
    df1 = pd.concat([df1,df2], axis = 0)

df_Oct = df1.copy()
```

```python
"""make the time in df_Jan as date time format, and remove Jan 5 date for df_Jan"""

df_Jan['time'] = pd.to_datetime(df_Jan['time'])
df_Jan = df_Jan[df_Jan['time'].dt.day != 5]
df_Jan = df_Jan[df_Jan['time'].dt.day != 30]
```

```python
"""make the time in df_Aug as date time format, and remove Aug 5, 24, 25 for df_Aug"""

df_Aug['time'] = pd.to_datetime(df_Aug['time'])
df_Aug = df_Aug[df_Aug['time'].dt.day != 1]
df_Aug = df_Aug[df_Aug['time'].dt.day != 24]
df_Aug = df_Aug[df_Aug['time'].dt.day != 25]
df_Aug = df_Aug[df_Aug['time'].dt.day != 30]
```

```python
df_Oct['time'] = pd.to_datetime(df_Oct['time'])
df_Oct = df_Oct[df_Oct['time'].dt.day != 17]
df_Oct = df_Oct[df_Oct['time'].dt.day != 29]
```

```python
print(len(df_Jan) / 9)
print(len(df_Aug) / 9)
print(len(df_Oct) / 9)
```

```python
Jan_diff = pd.read_csv('Jan_diff.csv')  
Aug_diff = pd.read_csv('Aug_diff.csv')
Oct_diff = pd.read_csv('Oct_diff.csv')
```

```python
Jan_diff = Jan_diff[node_list]
Aug_diff = Aug_diff[node_list]
Oct_diff = Oct_diff[node_list]
```

```python
Jan_day_ahead = pd.read_csv('Jan_day_ahead.csv')
Aug_day_ahead = pd.read_csv('Aug_day_ahead.csv')
Oct_day_ahead = pd.read_csv('Oct_day_ahead.csv')
```

```python
Jan_day_ahead = Jan_day_ahead[node_list]
Aug_day_ahead = Aug_day_ahead[node_list]
Oct_day_ahead = Oct_day_ahead[node_list]
```

```python
"""divide Jan_day_ahead by Jan_diff element-wise"""

Jan_return = Jan_diff / Jan_day_ahead
Aug_return = Aug_diff / Aug_day_ahead
Oct_return = Oct_diff / Oct_day_ahead
```

```python
"""save the return data to csv file"""

Jan_return.to_csv('Jan_return.csv')
Aug_return.to_csv('Aug_return.csv')
Oct_return.to_csv('Oct_return.csv')
```

```python
"""get rid of the prcp columns"""
df_Jan = df_Jan.drop(['prcp'], axis = 1)
df_Aug = df_Aug.drop(['prcp'], axis = 1)
df_Oct = df_Oct.drop(['prcp'], axis = 1)

"""normalize the data"""
df_Jan['temp'] = (df_Jan['temp'] - df_Jan['temp'].mean()) / df_Jan['temp'].std()   
df_Jan['rhum'] = (df_Jan['rhum'] - df_Jan['rhum'].mean()) / df_Jan['rhum'].std()
df_Jan['wspd'] = (df_Jan['wspd'] - df_Jan['wspd'].mean()) / df_Jan['wspd'].std()
df_Aug['temp'] = (df_Aug['temp'] - df_Aug['temp'].mean()) / df_Aug['temp'].std()
df_Aug['rhum'] = (df_Aug['rhum'] - df_Aug['rhum'].mean()) / df_Aug['rhum'].std()
df_Aug['wspd'] = (df_Aug['wspd'] - df_Aug['wspd'].mean()) / df_Aug['wspd'].std()
df_Oct['temp'] = (df_Oct['temp'] - df_Oct['temp'].mean()) / df_Oct['temp'].std()
df_Oct['rhum'] = (df_Oct['rhum'] - df_Oct['rhum'].mean()) / df_Oct['rhum'].std()
df_Oct['wspd'] = (df_Oct['wspd'] - df_Oct['wspd'].mean()) / df_Oct['wspd'].std()
```

```python
from sklearn.linear_model import LinearRegression
table_predict_return_Jan = pd.DataFrame(columns = ['Node_Name'])
for node in node_list:
    a = pd.concat([Jan_return[node], df_Jan[df_Jan['Node_Name'] == node]], axis = 1)
    temp_node = df_Jan[df_Jan['Node_Name'] == node].set_index('time')
    x = Jan_return[node].to_numpy()
    temp_node['return'] = x
    predict = []
    for i in range(len(temp_node) - 1):
        x = temp_node[['temp', 'rhum', 'wspd']].iloc[:i+1, :].to_numpy()
        y = temp_node['return'].iloc[:i+1].to_numpy()
        model = LinearRegression().fit(x, y.reshape(-1,1))
        new_x = temp_node[['temp', 'rhum', 'wspd']].iloc[i+1, :].to_numpy()
        number = model.predict(new_x.reshape(1, -1))
        predict.append(number[0][0])

    predict = np.array(predict)
    predict = np.insert(predict, 0, 0)
    table_predict_return_Jan[node] = predict
```

```python
table_predict_return_Jan = table_predict_return_Jan.drop(['Node_Name'], axis = 1)
table_predict_return_Jan
```
