```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib ipympl
```

```python
Jan_diff = pd.read_csv('Jan_diff.csv')
Aug_diff = pd.read_csv('Aug_diff.csv')
Oct_diff = pd.read_csv('Oct_diff.csv')
```

```python
"""read csv from url and convert to dataframe"""
geo_df = pd.read_csv('https://raw.githubusercontent.com/emunsing/CAISO-Scrapers/master/LMP%20Location%20Scraper/LMPLocations.csv')
```

```python
geo_df
```

```python
node_list = Jan_diff.columns.to_list()
node_lsit = node_list.remove('Unnamed: 0')
```

```python
"""find the rows with name values in the node_list"""
new_geo_df = geo_df[geo_df['name'].isin(node_list)]
new_geo_df
```

```python
new_node_list = new_geo_df['name'].to_list()
new_node_list
```

```python
"""use meteostat to get weather data given longtitude and latitude, data should be hourly"""
import meteostat
from meteostat import Stations, Daily, Hourly
import datetime

for node in new_node_list:
    a = new_geo_df[new_geo_df['name'] == node][['latitude', 'longitude']].to_numpy()
    lat = a[0][0]
    lon = a[0][1]
    start = datetime.datetime(2022, 1, 1)
    end = datetime.datetime(2022, 1, 30, 23, 59)
    stations = Stations()
    stations = stations.nearby(lat, lon)
    stations = stations.fetch(1)

    data = Hourly(stations, start, end)
    data = data.fetch()
    data = data[['temp', 'rhum', 'prcp', 'wspd']]
    data.to_csv(node + 'Jan' + '.csv')
    

```
